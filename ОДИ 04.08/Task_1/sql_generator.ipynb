```python
from dataclasses import field
import numpy
import pandas as pd
from mimesis import Generic, Field, Fieldset, Schema, Datetime
from mimesis.locales import Locale
from mimesis.plugins.pytest import mimesis
from mimesis.random import random
from datetime import datetime, timedelta
import pytest
```

```python
# Генерим зарегестрированных пользователей
user_fs = Fieldset(locale=Locale.EN, i=20)
user_df = pd.DataFrame.from_dict({
    "id": list(range(1, 21)),
    "name": user_fs("name"),
    "login": user_fs("username"),
    "password": user_fs("password")
})
user_df
```

```python
# Генерим темы
theme_dates = []
for _ in range(10):
    theme_dates.append(Datetime().formatted_datetime(fmt='%Y-%m-%d %H:%M:%S'))
    
theme_fs = Fieldset(locale=Locale.EN, i=10)
theme_df = pd.DataFrame.from_dict({
    "id": list(range(1, 11)),
    "name": theme_fs("word"),
    "text": theme_fs("text"),
    "creation_date": theme_dates,
    "creator_id": random.randints(10, 1, 21)
})
theme_df
```

```python
# Генерим сообщения
#Сгенерируем количество анонимных и неанонимных сообщений
random_anonym_messages_choices = random.choices([1, 0], k=50)
#Сгенерируем пользователя к каждому неанонимному сообщению
creator_id_with_anonym_messages = []
for boolean in random_anonym_messages_choices:
    if boolean != 1:
        creator_id_with_anonym_messages.append(random.randint(1, 21))
    else:
        creator_id_with_anonym_messages.append(0)

mess_dates = []
for _ in range(50):
    mess_dates.append(Datetime().formatted_datetime(fmt='%Y-%m-%d %H:%M:%S'))

message_fs = Fieldset(locale=Locale.EN, i=50)
message_df = pd.DataFrame.from_dict({
    "id": list(range(1, 51)),
    "text": message_fs("text"),
    "creation_date": mess_dates,
    "theme_id": random.randints(50, 1, 9),
    "creator_id": creator_id_with_anonym_messages,
})
message_df
```

Скриптом сгенерировать датасет содержащий для одного дня как минимум по 5 действий каждого типа. Для действия создания темы должно быть как минимум 2 случая возврата ошибки по причине отсутствия логина. Для действия написание сообщения должно быть примерно равное распределение залогиненых и незалогиненых пользователя. Скрипт для генерации должен сам выбирать число действий каждого типа используя рандомайзер но учитывая минимальные пороги.
Датасет должен содержать данные за месяц.


```python
# Генерим логи
date_start = datetime(2025, 2, 1)
log_id_counter = 1
logs = []
for day in range(28):
    date = date_start + timedelta(days = day)
    for action in ['first_visit', 'registration', 'login', 'logout',
    'create_theme', 'view_theme', 'delete_theme', 'create_message']:
        action_amount = random.randint(5, 10)
        
        #генерация создания темы
        if action == 'create_theme':
            for i in range(2):
                logs.append({
                    "id": log_id_counter,
                    "user_id": 0,
                    "action": 'create_theme',
                    "time": date + timedelta(minutes = random.randint(0, 1440)), 
                    "response": "Ошибка создания темы по причине отсутствия логина",
                    "anonym": 1
                })
                log_id_counter += 1
            for i in range(action_amount):
                logs.append({
                    "id": log_id_counter,
                    "user_id": random.randint(1, 21),
                    "action": 'create_theme',
                    "time": date + timedelta(minutes = random.randint(0, 1440)), 
                    "response": "Успешно!",
                    "anonym": 0
                })
                log_id_counter += 1
        
        #генерация только анонимных действий
        if action in ('first_visit', 'registration'):
            for i in range(action_amount):
                logs.append({
                    "id": log_id_counter,
                    "user_id": 0,
                    "action": action,
                    "time": date + timedelta(minutes = random.randint(0, 1440)),
                    "response": "Успешно!",
                    "anonym": 1
                })
                log_id_counter += 1
        
        #генерация действий только зарегестрированного пользователя
        if action in ('login', 'logout', 'delete_theme'):
            for i in range(action_amount):
                logs.append({
                    "id": log_id_counter,
                    "user_id": random.randint(1, 21),
                    "action": action,
                    "time": date + timedelta(minutes = random.randint(0, 1440)),
                    "response": "Успешно!",
                    "anonym": 0
                })
                log_id_counter += 1
        
        #генерация других действий
        if action in ('view_theme', 'create_message'):
            random_anonym_messages_choices = random.choices([1, 0], k = action_amount)
            user_ids = []
            for choice in random_anonym_messages_choices:
                if choice == 0:
                    user_ids.append(random.randint(1, 21))
                else:
                    user_ids.append(0)
                    
            for i in range(action_amount):
                logs.append({
                    "id": log_id_counter,
                    "user_id": user_ids[i],
                    "action": action,
                    "time": date + timedelta(minutes = random.randint(0, 1440)),
                    "response": "Успешно!",
                    "anonym": random_anonym_messages_choices[i]
                })
                log_id_counter += 1
        
logs_df = pd.DataFrame(logs)
logs_df         
```

```python
#Создаём файлы с данными
user_df.to_csv(header=True, sep = ';', index=False, path_or_buf='Task_1/data_files/user_data.csv', encoding='utf-8')
theme_df.to_csv(header=True, sep = ';', index=False, path_or_buf='Task_1/data_files/theme_data.csv', encoding='utf-8')
message_df.to_csv(header=True, sep = ';', index=False, path_or_buf='Task_1/data_files/messages_data.csv', encoding='utf-8')
logs_df.to_csv(header=True, sep = ';', index=False, path_or_buf='Task_1/data_files/logs_data.csv', encoding='utf-8')
```
